```python
import re
import os
```

```python
all_years = ['Educational Screen 1922','Educational Screen 1923','Educational Screen 1924',
             'Educational Screen 1925','Educational Screen 1926','Educational Screen 1927',
             'Educational Screen 1928','Educational Screen 1929','Educational Screen 1930',
             'Educational Screen 1931','Educational Screen 1932','Educational Screen 1933',
             'Educational Screen 1934','Educational Screen 1935','Educational Screen 1936',
             'Educational Screen 1937','Educational Screen 1938','Educational Screen 1939',
             'Educational Screen 1940','Educational Screen 1941','Educational Screen 1942',
             'Educational Screen 1943','Educational Screen 1944','Educational Screen 1945',
             'Educational Screen 1946','Educational Screen 1947','Educational Screen 1948',
             'Educational Screen 1949','Educational Screen 1950','Educational Screen 1951',
             'Educational Screen 1952','Educational Screen 1953','Educational Screen 1954',
             'Educational Screen 1955','Educational Screen 1956','Educational Screen 1957',
             'Educational Screen 1958','Educational Screen 1959','Educational Screen 1960',
             'Educational Screen 1961','Educational Screen 1962']

def clean_txt(doc):
    
    doc = re.sub('\n+', '\n', doc)
    doc = re.sub('\s\n', '\n', doc)
    doc = re.sub('-\n', '', doc)
    doc = re.sub('[^\S\r\n]{2,}', ' ', doc)
    doc = re.sub('\s-\s', '-', doc)

    return doc

for x in all_years:
    txtin = open('/home/dustin/Documents/Educational Screen/resources/' + x, 'r')
    read_txt = txtin.read()
    cleaned_txt = clean_txt(read_txt)
    txtout = open('/home/dustin/Documents/Educational Screen/resources/cleaned/' + x[19:23], 'w')
    txtout.write(cleaned_txt)
```

```python
def TwoWordSearch(text, word1, word2, wd):
    
    word_distance = wd

    w1 = word1
    w2 = word2

    pattern = "(" + w1 + "[\S\s]{0," + word_distance + "}" + w2 + ")|(" + w2 + "[\S\s]{0," + word_distance + "}" + w1 + ")"

    txt = open(text, 'r')
    rd = txt.read()

    matches = []
    for match in re.finditer(pattern, rd, flags=re.I + re.M):
        index = match.start()
        start = index - 1000
        stop = index + 1000
        full_match = rd[start:stop]
        matches.append(full_match)

    return matches

def OneWordSearch(text, word):

    txt = open(text, 'r')
    rd = txt.read()

    matches = []
    for match in re.finditer(word, rd, flags=re.I + re.M):
        index = match.start()
        start = index - 500
        stop = index + 500
        full_match = rd[start:stop]
        matches.append(full_match)

    return matches

def search_all_2word(names, words1, words2):

    filepaths = []
    for name in names:
        filepath = '/home/dustin/Documents/Educational Screen/resources/cleaned/' + name
        filepaths.append(filepath)

    for filepath in filepaths:
        for word1 in words1:
            for word2 in words2:
                index = 1
                results = TwoWordSearch(filepath, word1, word2, '500')
                for result in results:
                    txtout = open('/home/dustin/Documents/Educational Screen/results/' + filepath[59:64] + " " + word1 + " " + word2 + " " + str(index), 'w')
                    txtout.write(result)
                    index += 1

    return None

def search_all_1word(names, words):

    filepaths = []
    for name in names:
        filepath = '/home/dustin/Documents/Educational Screen/resources/cleaned/' + name
        filepaths.append(filepath)

    for filepath in filepaths:
        for word in words:
            index = 1
            results = OneWordSearch(filepath, word)
            for result in results:
                txtout = open('/home/dustin/Documents/Educational Screen/results/' + filepath[59:64] + " " + word + " " + str(index), 'w')
                txtout.write(result)
                index += 1

    return None

name_lst = os.listdir('/home/dustin/Documents/Educational Screen/resources/cleaned')
search_terms1 = [' plant', ' bean', 'root', ' tree ', ' flower', ' blossom', ' crop ']
search_terms2 = ['speed', 'grow']
search_terms3 = ['speed','stop motion', 'stop-motion', 'stopmotion', 'time lapse', 'time-lapse', 'timelapse', 'John Ott', 'John Nash Ott']

search_all_2word(name_lst, search_terms1, search_terms3)
# search_all_1word(name_lst, search_terms4)
```
